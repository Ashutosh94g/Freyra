# Freyra — T4-Optimised Colab Notebook

**Run cells in order.**  
Cell 2 installs packages and **automatically restarts the runtime** — that is expected.  
After the restart, skip Cell 1 & 2 and run **Cell 3 only**.


```python
# ── Cell 1 · Clone repo ─────────────────────────────────────────────────
import os

repo_dir = '/content/Freyra'
if not os.path.exists(repo_dir):
    os.system('git clone https://github.com/Ashutosh94g/Freyra.git ' + repo_dir)
else:
    print('Repo already cloned.')

os.chdir(repo_dir)
print('Working directory:', os.getcwd())
```

```python
# ── Cell 2 · Install dependencies (runtime restarts automatically) ────────────
import sys, subprocess, os
os.chdir('/content/Freyra')

# Proxy check: if numpy < 2.0 is present, install already succeeded in a
# previous run and we can skip straight to Cell 3.
try:
    import numpy as _np
    _already_done = int(_np.__version__.split('.')[0]) < 2
except Exception:
    _already_done = False

if _already_done:
    print(f'\u2705 Dependencies already installed (numpy={_np.__version__}). Skip to Cell 3.')
else:
    # Install CUDA torch first (Colab default may be CPU-only after reset)
    print('\U0001f4e6 Installing PyTorch 2.3.0 + CUDA 12.1 ...')
    subprocess.run(
        [sys.executable, '-m', 'pip', 'install', '-q',
         'torch==2.3.0', 'torchvision==0.18.0',
         '--extra-index-url', 'https://download.pytorch.org/whl/cu121'],
        check=True
    )
    # Install Fooocus requirements (most will say 'already satisfied')
    print('\U0001f4e6 Installing Fooocus requirements ...')
    subprocess.run(
        [sys.executable, '-m', 'pip', 'install', '-q', '-r', 'requirements_versions.txt'],
        check=True
    )
    print('\u2705 Done! Restarting runtime to pick up new packages ...')
    # Automatic restart — after this, skip to Cell 3
    import IPython
    IPython.Application.instance().kernel.do_shutdown(True)
```

```python
# ── Cell 3 · Launch (run after runtime restart) ──────────────────────────
#
#  --share               public Gradio URL
#  --always-high-vram    keep models on GPU (avoids CPU<->GPU swap)
#  --unet-in-fp8-e4m3fn  UNet VRAM: 6.5 GB -> 3.3 GB on T4
#  --preset realistic    realistic.json (no refiner)
#
import os
os.chdir('/content/Freyra')
os.environ['GRADIO_SERVER_PORT'] = '7865'

os.system(
    'python launch.py '
    '--share '
    '--always-high-vram '
    '--unet-in-fp8-e4m3fn '
    '--preset realistic'
)
```
